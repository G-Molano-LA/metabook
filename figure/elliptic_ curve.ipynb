```python
import numpy as np
# import matplotlib as mpl
# mpl.rcParams['figure.dpi'] = 400
# mpl.rcParams["figure.figsize"] = [10, 6]

import matplotlib.pyplot as plt
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

```python
def elliptic_curve(ax, a, b):
    y, x = np.ogrid[-5:5:200j, -5:5:200j]
    ax.contour(x.ravel(), y.ravel(), pow(y, 2) - pow(x, 3) - x * a - b, [0])
    ax.set_title("a="+str(a)+" ,b="+str(b))
    ax.grid()
    
figsize = (12,10)
nrows   = 4
ncols   = 4
astart  = -2
bstart  = -1

fig, axs = plt.subplots(nrows, ncols, figsize=figsize, constrained_layout=True)
for i in range(astart, astart+4):
    for j in range(bstart, bstart+4):
        elliptic_curve(axs[i+2,j+1],i,j)
        
    
plt.savefig('elliptic_curves.pdf', transparent=True, dpi=400)
plt.show()
```
